```python
directory  = "C://Users/Kers/Desktop/hanlun_dataset/videos"
def join(l):
    result = ""
    for item in l:
        result = result + item+ "/"
    return result[:-1]

# Loop all the video folder
# location: ./videos/
for data_dirname in os.listdir(directory):
    # skip file .DS_Store 
    if data_dirname  in ['.DS_Store', '._.DS_Store']:
        continue

    # date folders
    for skill_dirname in os.listdir(join([directory, data_dirname])):

        # skip file .DS_Store 
        if skill_dirname  in ['.DS_Store', '._.DS_Store']:
            continue

        # person name folders
        for person_name_dirname in os.listdir(join([directory, data_dirname, skill_dirname])):   

            # skip file .DS_Store 
            if person_name_dirname  in ['.DS_Store', '._.DS_Store']:
                continue    
            folder = (join([directory, data_dirname, skill_dirname]) + '/' + person_name_dirname + "/video1/frames/")
            imgs = []
            print(folder)
            for e in ['*.png', '*.jpg', '*.jpeg']: 
                imgs = imgs + glob.glob(folder + e)
            imgs_length = len(imgs)
            if imgs_length == 0:
                print("no imgs")
            else:
                print("Image number = ", imgs_length)



```
